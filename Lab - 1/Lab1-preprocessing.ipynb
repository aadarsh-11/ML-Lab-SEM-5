```python
# Library Imports

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Dataset Imports

data = pd.read_csv('Lab1-Data.csv')
data.head(10)
```

```python
# Viewing Dataset Statistically

data.describe()
```

```python
# dropping duplicate values

data = data.drop_duplicates()
data
```

```python
# Checking the null values

data.isnull()
```

```python
# Statistically checking the missing values

data.isnull().sum()
```

```python
data.dropna(how='any',inplace=True)
```

## Splitting the Data

```python
# Splitting dataset

X = data[['job_category', 'count', "percentage"]].values
Y = data['gender'].values
Z = data['race_ethnicity'].values
print(X)
print(Y)
print(Z)
```

## Categorical Data Encodding

```python
# Handling Categorical Data

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
ct = ColumnTransformer(transformers=[('enconder', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
# One-Hot Encoding Categorical Data

X = np.array(ct.fit_transform(X))
print(X)
```

```python
Y = le.fit_transform(Y)
print(Y)
```

```python
Z = le.fit_transform(Z)
print(Z)
```

## Splitting Test and Train Data

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting Test and Train Data on Y as dependent data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# Splitting Test and Train Data on Z as dependent data

X_train, X_test, Z_train, Z_test = train_test_split(X, Z, test_size=0.2, random_state=2)
```

```python
print(X_train)
print(Y_train)
print(Z_train)
print(X_test)
print(Y_test)
print(Z_test)
```
