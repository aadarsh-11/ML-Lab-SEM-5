```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
pred = km.fit_predict(df[['petal_length','petal_width']])
```

```python
x1['species'] = pred

x1.tail(20)
```

```python
df1 = x[x1.species==0]
df2 = x[x1.species==1]
df3 = x[x1.species==2]
```

```python
plt.scatter(df1['petal_length'],df1['petal_width'],color='blue')
plt.scatter(df2['petal_length'],df2['petal_width'],color='green')
plt.scatter(df3['petal_length'],df3['petal_width'],color='yellow')
#plt.scatter(km.cluster_centers_[:,2],km.cluster_centers_[:,3],color='purple',marker='+',label='centroid')
```

```python
sse = []
r1 = range(1,10)
for k in r1:
    km = KMeans(n_clusters=k)
    km.fit(x)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(r1,sse)
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.8,min_samples=14)
db.fit(x)
```

```python
cluster=df[db.labels_!=-1]
outlier=df[db.labels_==-1]
colors= db.labels_
ccluster=colors[colors!=-1]
ocluster='black'
```
